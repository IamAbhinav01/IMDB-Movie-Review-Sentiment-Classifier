```python
sent = ['Artificial intelligence is transforming the future of technology.',
'She enjoys reading books on rainy evenings.',
'He bought a new laptop for his college project.',
'The restaurant was crowded, but the food was excellent.',
'They traveled across Europe by train last summer.']
```

```python
sent
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=10000
```

```python
one_hot_rep = [one_hot(words,voc_size) for words in sent]
```

```python
one_hot_rep
```

```python
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
max_length = 10
embedded_docs = pad_sequences(one_hot_rep,padding='pre',maxlen=max_length)
```

```python
embedded_docs
```

```python
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=max_length))
model.compile('adam','mse')
model.summary()
```

```python
model.predict(embedded_docs)
```
